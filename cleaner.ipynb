# Data Cleaner for 2024 Data

```python
import pandas as pd
import re
```

```python
teams24 = pd.read_csv('./data/2024tourn.csv')
```

```python
teams24.head()
```

```python
teams24['SEED'] = teams24.apply(lambda row: row['TEAM'].split('\n')[1], axis=1)
teams24['TEAM'] = teams24.apply(lambda row: row['TEAM'].split('\n')[0], axis=1)

# for col in teams24.columns:
#     if col != 'TEAM':
#         teams24[col] = pd.to_numeric(teams24[col])

#
```

```python
teams24.head()
```

```python
for col in teams24.columns:
    try:
        teams24[col] = teams24.apply(lambda row: row[col].split('\n')[0], axis=1)
    except AttributeError:
        pass
```

```python
teams24.head()
```

```python
teams24['W'] = teams24.apply(lambda row: row['REC'].split('-')[0], axis=1)
teams24['WIN_PCT'] = teams24['W'].astype(int)/teams24['G']
teams24.drop(['G', 'W'], inplace=True, axis=1)
teams24.rename(columns={"EFG%": "EFG_O", "EFGD%": "EFG_D", "2P%" : "2P_O", "2P%D" : "2P_D", "3P%" : "3P_O", "3P%D" : "3P_D", "ADJ T." : "ADJ_T"}, inplace=True)
teams24.head()
```

```python
teams24.drop(['REC', 'SEED'], inplace=True, axis=1)
```

```python
teams24.drop('RK', inplace=True, axis=1)
```

```python
teams24
```

```python
teams24['YEAR'] = 2024
```

```python
teams24.head()
```

```python
teams24.to_csv('./data/2024_formatted.csv')
```
